```python
# Import dependencies

import pandas as pd
```

```python
# Read CSV file 

epl_csv = pd.read_csv("Resources/EPL_season_1819.csv")

epl_csv.head()
```

```python
# Collect information we need

epl_df = epl_csv[["Date", "HomeTeam", "AwayTeam", "FTHG", "FTAG", "FTR"]]

epl_df.head()
```

```python
# Rename columns

renamed_epl_df = epl_df.rename(columns = {"HomeTeam": "Home Team",
                                  "AwayTeam": "Away Team",
                                  "FTHG": "Home Team Goals",
                                  "FTAG": "Away Team Goals",
                                  "FTR": "Game Result"})

renamed_epl_df.head()
```

```python
# Sort data by "Home Team"

manchester_home_games = renamed_epl_df.sort_values("Home Team", ascending = True)

manchester_home_games.head()
```

```python
# Sort data by "Away Team"

manchester_away_games = renamed_epl_df.sort_values("Away Team", ascending = True)
manchester_away_games.head()
```

```python
# Show only Manchester City games played in Manchester and sort by "Date"

mancity_home_games = manchester_home_games.loc[manchester_home_games["Home Team"] == "Man City", :]

mancity_home_games.sort_values(["Date"])
```

```python
# Show only Manchester City games played away Manchester and sort by "Date"

mancity_away_games = manchester_away_games.loc[manchester_away_games["Away Team"] == "Man City", :]

mancity_away_games.sort_values(["Date"])
```

```python
# Show only Manchester United games played in Manchester and sort by "Date"

manunited_home_games = manchester_home_games.loc[manchester_home_games["Home Team"] == "Man United", :]

manunited_home_games.sort_values(["Date"])
```

```python
# Show only Manchester United games played away Manchester and sort by "Date"

manunited_away_games = manchester_away_games.loc[manchester_away_games["Away Team"] == "Man United", :]

manunited_away_games.sort_values(["Date"])
```

```python
# Combine Manchester City and Manchester United home games in one Data Frame 
# Show only games played in Auguest 2018

combined_manchester_home_games = pd.merge(mancity_home_games, manunited_home_games, how = "outer")
combined_manchester_home_games.sort_values(["Date"]).head()

home_auguest_manchester_games = combined_manchester_home_games.sort_values(["Date"]).head(3)
home_auguest_manchester_games.reset_index(drop = True)

```

```python
# Combine Manchester City and Manchester United away games in one Data Frame 
# Show only games played in Auguest 2018

combined_manchester_away_games = pd.merge(mancity_away_games, manunited_away_games, how = "outer")
combined_manchester_away_games.sort_values(["Date"]).head()

away_auguest_manchester_games = combined_manchester_away_games.sort_values(["Date"]).head(3)
away_auguest_manchester_games.reset_index(drop = True)
```

```python
manchester_crime_data = pd.read_csv("Resources/final_manchester_crime_data.csv")
manchester_crime_data.head()
```

```python
manchester_crime_dataset = manchester_crime_data[["CrimeTS", "Location", "Type", "latitude", "longitude"]]
manchester_crime_dataset.head()
```

```python
manchester_crime_dataset['Date'] = pd.to_datetime(manchester_crime_dataset['CrimeTS']).dt.normalize()
```

```python
manchester_crime_dataset = manchester_crime_dataset[["Date", "Location", "Type", "longitude", "latitude"]]
manchester_crime_dataset.rename(columns = {"longitude": "Latitude", "latitude": "Longitude"}).head()
```

```python
first_away_game = manchester_crime_dataset.loc[manchester_crime_dataset["Date"] == "2018-08-12"].reset_index()
del first_away_game["index"]

second_away_game = manchester_crime_dataset.loc[manchester_crime_dataset["Date"] == "2018-08-19"].reset_index()
del second_away_game["index"]

third_away_game = manchester_crime_dataset.loc[manchester_crime_dataset["Date"] == "2018-08-25"].reset_index()
del third_away_game["index"]
```

```python
import gmaps

API_KEY = 'Your Code'

gmaps.configure(api_key=API_KEY)

stadium_locations = [(53.4631, -2.2913), (53.4831, -2.2004)]
crime_locations_first = first_away_game[["longitude", "latitude"]].astype(float)

fig = gmaps.figure(map_type="TERRAIN")

heatmap_layer = gmaps.heatmap_layer(crime_locations_first)
fig.add_layer(heatmap_layer)


crime_layer = gmaps.symbol_layer(
    crime_locations_first, fill_color='white',
    stroke_color='black', scale=3, 
)

stadium_markers = gmaps.marker_layer(stadium_locations, info_box_content=[f"Stadium location: {stadium}" for stadium in stadium_locations])

fig.add_layer(crime_layer)
fig.add_layer(stadium_markers)
fig
```
